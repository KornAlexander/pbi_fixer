## 📋 Abstract

The **Power BI Fixer** is an interactive development environment that runs natively inside Microsoft Fabric Notebooks. It combines **Semantic Link**, **ipywidgets**, and **TOM (Tabular Object Model)** to provide a comprehensive tool for scanning, exploring, fixing, and documenting Power BI reports and semantic models — all from a single notebook cell.

### The Problem It Solves

Power BI developers face a fragmented workflow:
- **Report issues** (pie charts, missing data labels, page sizes) require manual visual-by-visual fixes in Power BI Desktop
- **Model best practices** (missing descriptions, floating point types, foreign key visibility) need Tabular Editor or manual scripting
- **No single tool** lets you scan a report AND its underlying model, see the results side-by-side, and fix everything in one workflow
- **PBIR format adoption** is slow because upgrade tooling is scattered

The PBI Fixer addresses all of these by providing **12+ interactive tabs**, **17 report fixers**, **25+ semantic model fixers**, and a unified scan-and-fix workflow — running entirely in a Fabric Notebook.
> **~17,000 lines of new Python code** across 5 UI & tab modules (9,439 lines — incl. Fix All, Model Explorer, Report Explorer, Prototype, Translations, Model Diagram, About), 15 report fixers (3,038 lines), 4 report helpers (593 lines), and 33 semantic model fixers (3,797 lines) — all written from scratch for this project.

## 📊 Feature Overview — 12 Interactive Tabs

The PBI Fixer organizes all capabilities into tabbed panels. By default, 4 tabs load for **fast startup**. Set `all_tabs=True` to unlock all 12.

| 9 | **Report BPA** | 📄 | Best Practice Analyzer for reports — detect anti-patterns and fix them |

| 11 | **Prototype** | ✏️ | Generate visual report prototypes as SVG/Excalidraw with page navigation mapping |

## 🔍 Example: Calling Individual Fixers via Python

While the interactive UI is the recommended way to use the PBI Fixer, you can also call fixers programmatically.  
This is useful for CI/CD pipelines, batch processing, or custom automation scripts.

````python
import sempy_labs as labs

# --- Report Fixers (require PBIR format) ---

# Fix all pie charts → bar charts
labs.fix_piecharts(report="My Report", workspace="My Workspace")

# Fix bar chart formatting (add data labels, remove gridlines)
labs.fix_barcharts(report="My Report", workspace="My Workspace")

# Upgrade report to PBIR format
labs.fix_upgrade_to_pbir(report="My Report", workspace="My Workspace")
````

````python
import sempy_labs as labs

# --- Semantic Model Fixers (via XMLA / TOM) ---

# Hide all foreign key columns
labs.fix_hide_foreign_keys(dataset="My Model", workspace="My Workspace")

# Auto-create measures from SummarizeBy columns
labs.add_measures_from_columns(dataset="My Model", workspace="My Workspace")

# Add PY time intelligence measures
labs.add_py_measures(dataset="My Model", workspace="My Workspace")

# Prepare model for Copilot (descriptions, categories, formatting)
labs.fix_prep_for_ai(dataset="My Model", workspace="My Workspace")
````

## 🛠️ Architecture Overview

### How It Works

```
┌─────────────────────────────────────────────────────────────────┐
│                    Fabric Notebook (ipywidgets)                 │
├─────────────────────────────────────────────────────────────────┤
│  pbi_fixer()                                                    │
│  ├── Fix All Tab ──────── Runs all fixers in sequence          │
│  ├── Model Explorer ───── sempy.fabric (REST) + TOM (XMLA)    │
│  ├── Report Explorer ──── connect_report() → PBIR JSON        │
│  ├── Perspectives ─────── TOM: Model.Perspectives              │
│  ├── Translations ─────── TOM + Azure AI Translator            │
│  ├── Memory Analyzer ──── Vertipaq DMVs via XMLA              │
│  ├── Model BPA ────────── Rule engine + TOM fixers             │
│  ├── Report BPA ───────── Rule engine + PBIR fixers             │
│  ├── Delta Analyzer ───── Delta Lake metadata                  │
│  ├── Prototype ────────── SVG/Excalidraw generation            │
│  └── Model Diagram ────── Relationship visualization           │
├─────────────────────────────────────────────────────────────────┤
│  Dependencies:                                                  │
│  • sempy (Semantic Link) — Microsoft's official library        │
│  • sempy_labs — Community extensions by Michael Kovalsky       │
│  • ipywidgets — Interactive widget framework                   │
│  • .NET interop — TOM (Tabular Object Model) via Python       │
└─────────────────────────────────────────────────────────────────┘
```

### Key Integration Points

| Component | Technology | Purpose |
|-----------|------------|--------|
| **Semantic Link** | `sempy.fabric` | REST API for listing workspaces, items, models, reports |
| **TOM** | `.NET interop` | Read/write semantic model objects (measures, columns, tables, relationships) |
| **PBIR** | `connect_report()` | Parse and modify report JSON definitions |
| **ipywidgets** | `SelectMultiple`, `HTML`, `Tab` | Interactive UI components in the notebook |
| **Vertipaq DMVs** | XMLA queries | Memory and storage analysis |
| **Azure AI** | Translator API | Auto-translate object names for multi-language support |
